````python
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib
matplotlib.style.use('ggplot')

np.random.seed(1)
mu , sigma = 5 , 2
data = np.round(np.random.normal(mu,sigma,100))
plt.hist(data,bins=10,range=(0,10),edgecolor='black')
plt.show()
````

Strat Statiscal information

````python
#mean
mean=np.mean(data)
mean
````

````python
#Standart deviation
stdisv = np.std(data)
stdisv
````

````python
#Mediana
med = np.median(data)
med
````

````python
#Central Tendency

#mode

moda = stats.mode(data)

print("Resulta de los datos es: mediana={}, desviacion std = {}, media = {}".format(mean,stdisv,med))
print('')
print("Valor de la moda = {} y el count={}".format(moda.mode[0],moda.count[0]))
````

````python
# Spread of data == |max data - min data|
#Range
np.ptp(data)
````

````python
#Variance 
np.var(data)
````

Standar Error of the Mean:  Deviation of the mean if several measurements are performed.

````python
stats.sem(data)
````
