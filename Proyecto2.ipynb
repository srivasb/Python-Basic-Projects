```python
path = '../vol/datasets/peliculas.csv'
movies = pd.read_csv(path)
movies.head()
```

### Merging to DataFrames

```python
#Select correct columns
financials.columns
```

```python
financials = financials[['movie_title' , 'production_budget', 'worldwide_gross' ]]
financials
```

```python
financials.shape , movies.shape
```

```python
#Separate num columns from obtect columns

num = (movies.dtypes == float) | (movies.dtypes == 'int64') 
movies_num_index = [index for index in num.index if num[index]] #index are the names of columns in this boolean mask
movies_num = movies[movies_num_index]
movies_num.head(10)
```

##### Because both df must have same key, i.e., share a feature

```python
#Add a Feature: Movie_title (Column) to movies_num
#Concatenar function
#To concatenate a column: axis = 1

movies_num = pd.concat([movies_num, movies['movie_title']], axis = 1) 
```

### Merge

```python
#Merge two df by the KEY and keep the index length of the LEFT df

key = 'movie_title'
left_df_length = 'left'

movies_v2 = pd.merge(financials , movies_num , on = key , how = left_df_length)
```

```python
movies_v2.shape, financials.shape, movies_num.shape
```

## Eliminating NaN and Null values using ScikitLearn

```python
#Library to impute / assign / change / erase the non-desier values

from sklearn.impute import SimpleImputer
imputer = SimpleImputer( missing_values = np.nan , strategy = 'mean' )
```

```python
movies_v2
```

#####  Eliminate columns to work only with numerical values

```python
movies_v2 = movies_v2.drop('movie_title', axis = 1)
movies_v2 = movies_v2.drop('duration', axis = 1)
movies_v2
```

```python
movies_v2.values
```

```python
#Getting the values in an array and applying imputer object

values = imputer.fit_transform(movies_v2)
values, values.shape
```

```python
#Transfor array into df

x = pd.DataFrame(values)
```

```python
x.columns = movies_v2.columns
```

```python
x.index = movies_v2.index
```

```python
x.head(10)
```

##### Save Clean df into a csv

```python
x.to_csv('../vol/intermediate_result/x.csv', index = False)
```
